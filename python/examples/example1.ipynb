### Uso de `SciPy` (Fundamental algorithms for scientific computing in Python)

Ejemplo de una integral con Simpson 1/3 de la funcion:

$$
    y = cos(x)
$$

Entre 0 y $\frac{\pi}{2}$

```python
x = np.linspace(0, np.pi / 2, 1000)

y = np.cos(x)

integral = simpson(y, x = x)
```

```python
integral
```

El valor real de la integral:

$$
    \int_{0}^{\pi / 2} cos(x) \;\; dx
$$

$$
    sin(x) |_{0}^{\pi / 2}
$$

$$
    sin(\pi / 2) - sin(0) = sin(\pi / 2) = 1
$$

### Crear graficos con matplotlib

```python
x = np.linspace(0, 6 * np.pi, 1000)

y = np.cos(x)

plt.plot(x, y)

plt.xlim(0, 6 * np.pi)
plt.ylim(-1, 1)

plt.xlabel("X")
plt.ylabel("Y")

plt.title("Gr√°fico");
```
